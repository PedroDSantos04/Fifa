```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
```

```python
df = pd.read_csv('fifa.csv')
df.head(0)
```

```python
df_agg = df.groupby(['Age'], as_index=False)[['Age', 'Overall']].mean()
df_agg
```

```python
df1 = df.groupby(['Age'], as_index=False)[['Age']].count()
df1
```

```python
plt.figure(figsize=(12,16)).set_figheight(4)
sns.scatterplot(x = 'Age', y = 'Overall', hue = df1['Age'], palette = 'viridis', data = df_agg)
plt.title('MÃ©dia de Overall por idade')
plt.xticks(df_agg['Age'][::1])
plt.legend()
plt.xlabel('Idade')
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(12,16)).set_figheight(4)
color = ['black', 'grey', 'navy']
plt.bar(x = 'Age', height=  'Overall', data = df_agg, color=color)
plt.ylabel('MÃ©dia de Overall')
plt.xlabel('Idade')
plt.title('MÃ©dia de Overall por Idade')
plt.xticks(df_agg['Age'][::1])
plt.show()
```
